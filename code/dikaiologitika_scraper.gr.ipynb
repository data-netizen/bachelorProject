2) 'REQUESTS' METHOD & BEAUTIFUL SOUP

```python
artic_lst= []


for i in range(60, 160):     #define how many pages we want to get

    url = 'https://www.dikaiologitika.gr/eidhseis/ergasiaka?start=' + str(i)
    response = requests.get(url)
    doc = soup(response.text, 'html.parser')
    articles = doc.find_all('div', {'class': 'catItemBody'}) 
    for article in articles:
        title = article.find('h3', {'class': 'catItemTitle'})
        if title:
           title_text = title.text
        else:
           title_text = ''

        summary = article.find('div', {'class':'catItemIntroText'})
        if summary:
           summ_text = summary.text
        else:
           summ_text = ''

        time = article.find('span', {'class': 'catItemDateCreated'})
        if time:
           time_text = time.text
        else:
           time_text = ''
    

        data = {'Title': title_text, 'Summary': summ_text, 'Time': time_text, 'site': 'dikaiologitika.gr'}
        artic_lst.append(data)

               
                
          
      #find the summary(from the whole article tag, extract the sub-tags)
     
  
         
      #find the time(from the whole article tag, extract the-sub tags)
        

      #create dictionary
              
      #append the dictionary into the list you created above
             
```
