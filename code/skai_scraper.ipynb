```python
#import libraries
import requests                            
from bs4 import BeautifulSoup as soup  
import pandas as pd
import matplotlib.pyplot as plt  
% matplotlib inline
import numpy as np    
```

```python
# extract  articles with the keyword by pulling them out of the HTML code  
workbill_keyword= []

for i in range(0,9):

      url = 'https://www.skai.gr/s/%CE%B5%CF%81%CE%B3%CE%B1%CF%83%CE%B9%CE%B1%CE%BA%CF%8C%20%CE%BD%CE%BF%CE%BC%CE%BF%CF%83%CF%87%CE%AD%CE%B4%CE%B9%CE%BF?page=' + str(i)
      response = requests.get(url)
      doc = soup(response.text, 'html.parser')
      #find the 'big' tag of the article
      articles = doc.find_all('div' , {'class': 'cmnArticleTitlePad'})
      for article in articles:
          #find the subtags
          title = article.find('a', {'class': 'title mainLink'})
          if title:
             title_text = title.text
          else:
              title_text = ""
          
          summary = article.find('div', {'class': 'lead'})
          if summary:
             summary_text = summary.text
          else:
             summary_text = ""

          time = article.find('div', {'class': 'date'})
          if time:
             time_text = time.text
          else:
             time_text = ''
                  



          data = {'Title': title_text, 'Summary': summary_text, 'Time': time_text}
          #append the extracted data to the list we had created
          workbill_keyword.append(data)
      
```

```python
df = pd.DataFrame(workbill_keyword)
df
```
