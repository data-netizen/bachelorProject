```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
import numpy as np
```

```python
#Use 'requests' method to get the web page and beautiful soup to parse HTML
artic_lst= []

for i in range(1,29):       #define how many pages we want to get
    url = 'https://www.news247.gr/politiki/?pages=' + str(i)
    response = requests.get(url)
    doc = soup(response.text, 'html.parser')
  #find the 'big' tag of the article
    articles = doc.find_all('article' , {'class': 'teaser__article'})

  #find the subtags of the article using a for loop
    for article in articles:
          #get the title
          title = article.find('h3', {'class': 'article__title bold'})

          if title:
                title_text = title.text
          else:
                title_text = ''
          #get the summary
          summary = article.find('p', {'class': 'article__leadtext'})
          if summary:
             summary_text = summary.text
          else:
             summary_text = ''
          #get the date
          time = article.find('time' )
          if time:
             time_text = time['datetime']
          else:
             time_text = ''



          data = {'Titles': title_text , 'Summary': summary_text , 'Time': time_text}
          artic_lst.append(data)



```

```python
df = pd.DataFrame(artic_lst)
df
```

```python
pd.set_option('display.max_rows', df.shape[0]+1)
print(df)
```
