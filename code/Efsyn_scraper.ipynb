```python
import requests                            
from bs4 import BeautifulSoup as soup  
import pandas as pd
import matplotlib.pyplot as plt  
% matplotlib inline
import numpy as np    
```

```python
# pulling  articles with the keyword out of HTML code of the website
workbill_keyword= []

for i in range(0,160):

      url = 'https://www.efsyn.gr/search?created%5B0%5D=&created%5B1%5D=&field_anonymous_author=&field_author=All&field_category=All&keywords=%CE%B5%CF%81%CE%B3%CE%B1%CF%83%CE%B9%CE%B1%CE%BA%CF%8C%20&sort_by=created&page=' + str(i)
      response = requests.get(url)
      doc = soup(response.text, 'html.parser')
      #find the 'big' tag of the article
      articles = doc.find_all('article' , {'class': 'default-teaser__article'}) 
      for article in articles:
          # find the subtag of the title
          title = article.find('a', {'class': "full-link"})
          if title:
              title_txt = title.text  
          else:
              title_txt = ''
          # find the subtag of the summary
          summary = article.find('div', {'class': 'default-teaser__summary'})
          if summary:
             summary_txt = summary.text
          else:
             summary_txt = summary.text
          # find the subtag of the time
          time = article.find('time')
          if time:
             time_txt = time.text
          else:
             time_txt = ' '

          data = {'Title': title_txt, 'Summary': summary_txt, 'Time': time_txt, 'Site': 'skai.gr' }
          workbill_keyword.append(data)
      
```

```python
#create the dataframe
df = pd.DataFrame(workbill_keyword)
df
```

```python
#show all the articles
pd.set_option('display.max_rows', df.shape[0]+1)
print(df)
```
